```python
last_year = dt.date(2017, 8, 23) - dt.timedelta(days=365)
print(last_year)
```

```python
PrecipData = session.query(Measurement.date, Measurement.prcp).\
    filter(Measurement.date > last_year).\
    order_by(Measurement.date).all()
```

```python
#put into a data fram
PrecipData_df=pd.DataFrame(PrecipData)
PrecipData_df.head()
```

```python
PrecipData_df.plot('date', 'prcp')
plt.xlabel("Date")
plt.ylabel("Rain in Inches")
plt.legend(["Precipitation"])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
ActiveStations=session.query(Measurement.station,\
func.count(Measurement.tobs)).group_by(Measurement.station)\
.order_by(func.count(Measurement.tobs).desc()).all()
print(ActiveStations)
```

```python
MostActive=ActiveStations[0][0]
print(MostActive)
```

```python
tempObser= session.query(Measurement.station, Measurement.date, Measurement.tobs).\
    filter(Measurement.station == MostActive).\
    filter(Measurement.date > last_year).\
    order_by(Measurement.date).all()
```

```python
temp_df=pd.DataFrame(tempObser)
plt.hist(temp_df['tobs'],12)
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.legend(["tobs"])
plt.show()
```
